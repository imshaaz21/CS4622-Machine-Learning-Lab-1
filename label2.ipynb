##Label 03

```python
train2 = train[FEATURES + ['label_2']].dropna()
valid2 = valid[FEATURES + ['label_2']]
```

```python
train2.head()
```

```python
from sklearn import svm
```

```python
from sklearn.preprocessing import StandardScaler
```

##  without any feature engineering

```python
train2 = train2.dropna()
valid2 = valid2.dropna()
```

```python
scaler = StandardScaler()
```

```python
_x_train2 = train2.drop('label_2',axis=1)
y_train2 = train2['label_2']

_x_valid2 = valid2.drop('label_2',axis=1)
y_valid2 = valid2['label_2']
```

```python
x_train2 = pd.DataFrame(scaler.fit_transform(_x_train2),columns=FEATURES)
x_valid2 = pd.DataFrame(scaler.transform(_x_valid2),columns=FEATURES)
```

```python
clf = svm.SVC(kernel='linear')
clf.fit(x_train2,y_train2)
```

```python
y_pred = clf.predict(x_valid2)
```

```python
from sklearn import metrics
```

```python
def find_accuracy(Y_pred,Y_valid):
  print(f"accuracy_score {metrics.accuracy_score(Y_valid,Y_pred)}")
  print(f"precision_score {metrics.precision_score(Y_valid,Y_pred,average='weighted')}")
  print(f"recall_score {metrics.recall_score(Y_valid,Y_pred,average='weighted')}")
```

```python
find_accuracy(y_pred,y_valid2)
```

## Feature selection

### Mutual Information Feature selection

```python
from sklearn.feature_selection import SelectKBest, mutual_info_classif
import numpy as np
```

```python
for k in [50,150,230]:
  mutual_info_selector = SelectKBest(mutual_info_classif, k=k)
  x_train_mutual_info = mutual_info_selector.fit_transform(x_train2, y_train2)
  print(f"Selected featured: ", x_train_mutual_info.shape[1])

  clf = svm.SVC(kernel='linear')
  clf.fit(x_train_mutual_info,y_train2)
  y_pred = clf.predict(mutual_info_selector.transform(x_valid2))

  print(f"Accuracies for k={k}: ")
  find_accuracy(y_pred,y_valid2)
```

k = 230

```python
mutual_info_selector = SelectKBest(mutual_info_classif, k=200)
x_train_mutual_info = mutual_info_selector.fit_transform(x_train2, y_train2)
print(f"Selected featured: ", x_train_mutual_info.shape[1])

clf = svm.SVC(kernel='linear')
clf.fit(x_train_mutual_info,y_train2)
y_pred = clf.predict(mutual_info_selector.transform(x_valid2))

print(f"Accuracies for k={200}: ")
find_accuracy(y_pred,y_valid2)
```

### PCA

```python
from sklearn.decomposition import PCA
```

```python
for n in [0.95,0.99]:
  pca = PCA(n_components=n, svd_solver='full')
  pca.fit(x_train2)
  x_train_pca = pd.DataFrame(pca.transform(x_train2))
  x_valid_pca = pd.DataFrame(pca.transform(x_valid2))
  print(f"Selected featured: ", x_train_pca.shape[1])

  clf = svm.SVC(kernel='linear')
  clf.fit(x_train_pca,y_train2)
  y_pred = clf.predict(x_valid_pca)
  print(f"Accuracies for n_components={n}: ")
  find_accuracy(y_pred,y_valid2)
```

### Mutual Information Feature selection + PCA k=230, n_components=0.99

```python
mutual_info_selector = SelectKBest(mutual_info_classif, k=230)
x_train_mutual_info = mutual_info_selector.fit_transform(x_train2, y_train2)
print(f"Selected featured for k=230: ", x_train_mutual_info.shape[1])

pca = PCA(n_components=0.99, svd_solver='full')
pca.fit(x_train2)
x_train_pca = pd.DataFrame(pca.transform(x_train2))
x_valid_pca = pd.DataFrame(pca.transform(x_valid2))
print(f"Selected featured: ", x_train_pca.shape[1])

clf_selected = svm.SVC(kernel='linear')
clf_selected.fit(x_train_pca,y_train2)
y_pred = clf_selected.predict(x_valid_pca)
print(f"Accuracies for n_components={0.99}: ")
find_accuracy(y_pred,y_valid2)
```

### Correlation Co-efficient

```python
corr_matrix3 = x_train2.corr()
```

```python
upper_triangle2 = corr_matrix3.where(np.triu(np.ones(corr_matrix3.shape), k=1).astype(np.bool))
correlated_features_to_drop2 = [column for column in upper_triangle2.columns if any(upper_triangle2[column] > 0.5)]
```

```python
print("Correlated features to drop:", correlated_features_to_drop2)
print("Number of features to drop:", len(correlated_features_to_drop2))
print("Number of features left:", 256 - len(correlated_features_to_drop2))
```

```python
clf_cc = svm.SVC(kernel='linear')
clf_cc.fit(x_train2[x_train2.columns.difference(correlated_features_to_drop2)],y_train2)
y_pred = clf_cc.predict(x_valid2[x_valid2.columns.difference(correlated_features_to_drop2)])
print(f"Accuracy: ")
find_accuracy(y_pred,y_valid2)
```
